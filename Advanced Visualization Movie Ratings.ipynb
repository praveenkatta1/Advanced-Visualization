# Kernal Density Estimate Plot

```python
sns.kdeplot(ratings['Critic Ratings'], label='Critic Ratings')
sns.kdeplot(ratings['Audience Ratings'], label='Audience Ratings')


plt.legend()

plt.show()
```

```python
sns.kdeplot(ratings['Audience Ratings'], shade=True, shade_lowest=False, cmap='Blues', label='Audience Ratings')
sns.kdeplot(ratings['Critic Ratings'], shade=True, shade_lowest=False, cmap='Reds', label='Critic Ratings')


plt.legend()


plt.show()
```
